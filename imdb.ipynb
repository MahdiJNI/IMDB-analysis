converting vote counts from numbers with 'K' and 'M' suffixes which imply to thousand and milliom to numerical values with those number units.
for instance :
3.1m, which is three million and one hundered thousand, is converted to 3100000
89k, which is eighty nine thousand, is converted to 89000

```python
to_modify = df['vote_count']
for index, row in to_modify.items():
    if 'm' in row:
        to_modify.loc[index] = int(float(to_modify.loc[index].split('m')[0]) * 1000000)
    elif 'k' in row:
        to_modify.loc[index] = int(float(to_modify.loc[index].split('k')[0]) * 1000)
df['vote_count'] = to_modify
```

```python
df['vote_count'] = df['vote_count'].astype(int)
```

then here, 
values in duration column with hours and minutes all together will be converted to minutes
for instance :
2h 22m, which is two hours and 22 minutes will become 122 minutes.
the ones that are only minutes will remain the same, 
for instance :
48m, will remain 48 but the only change is that it's just gonna change to a numeric value, which is 48

```python
edition = df['duration']
for index, row in edition.items():
    if 'm' in row and 'h' in row:
        x1 = edition.loc[index].split(r'h')
        x2 = int(x1[1].split('m')[0]) + int(x1[0])*60
        edition.loc[index] = x2
    elif 'h' in row and 'm' not in row:
        edition.loc[index] = int(edition.loc[index].split(r'h')[0]) * 60
    elif 'm' in row and 'h' not in row:
        edition.loc[index] = int(edition.loc[index].split(r'm')[0])
df['duration'] = edition
```

```python
df['duration'] = df['duration'].astype(int)
```

```python
df.rename(columns={'duration':'duration_in_minute'}, inplace=True)
```

removing outliers, just for       release year    and    imdb rate

we don't remove outliers in the duration column, vote count and box office
because it is not reasonable to do it. if a movie is 10 hours long it's not an outlier because it's just longer than the rest, though, the formula will detect it as an outlier, but removing that case will just cause Data Loss.
even it box office, a movie sales can grow significantly high while the others can't or don't. if the formula detects it as an outlier, we don't remove that Data because it's just a movie that was much much more successful than the rest !!

In the cases of release year & imdb rate it is totally different.
because if we have a value that is minus 2 (-2) or a movie with 11 out of 10 rate (11/10),
these values are definitely outliers because they exceed the solid limits, 11 out of 10 is reasonable here, so it is an absolute outlier
Or
a movie with release year like 1600 or 2080 is also an outlier
because there was no camera in 1600 to record a movie or the year 2080 has not arrived yet !!
so these values are also considered as outliers

```python
q1 = df.release_year.quantile(0.25)
q3 = df.release_year.quantile(0.75)
iqr = q3 - q1
lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr
df[(df.release_year<lower_limit)|(df.release_year>upper_limit)]
```

here, the formula detects a few movies with imdb rates around 8.9 to 9.3 as outliers,
as you see they are not outliers, they're just a few movies that are more successful than the rest !!

```python
q1 = df.imdb_rate.quantile(0.25)
q3 = df.imdb_rate.quantile(0.75)
iqr = q3 - q1
lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr
df[(df.imdb_rate<lower_limit)|(df.imdb_rate>upper_limit)]
```
